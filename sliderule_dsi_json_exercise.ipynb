## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
#data is a list of dictionaries, 4 key value pairs each .
#state and shortname are keys to string values
#info is the key to a dictionary with one key value pair, string value
#counties is the key to a list of dictionaries with two key value pairs each, keys to string and int values
```

```python
json_normalize(data, 'counties', 'info')
#apparently a list can be put straight into the table, but not a dictionary
```

```python
json_normalize(data, 'counties', [['info','governor']])
#this would not work unless info, governor was in double brackets
#it also would not work without the counties list in front, shown below
```

```python
json_normalize(data, 'counties', ['state', 'shortname'])
#json_normalize(data, 'counties', 'state', 'shortname')
#I don't understand why this commented out line doesn't work
#they need the bracket to appear as separate columns
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```
